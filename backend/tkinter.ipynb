```python
import tkinter as tk
from tkinter import ttk, Text, messagebox
import json
import os

# ============================
# Clase Nodo y ColaDob
# ============================
class Nodo:
    def _init_(self, data=None):
        self.next = None
        self.data = data
        self.bef = None

class ColaDob:
    def _init_(self, limit=9):
        self.limite = limit
        self.front = None
        self.rear = None
        self.size = 0

    def isEmpty(self):
        return self.size <= 0

    def isFull(self):
        return self.size >= self.limite

    def enColarD(self, data):
        nodo = Nodo(data)
        if self.isFull():
            print("Cola llena")
            return

        if self.isEmpty():
            self.front = self.rear = nodo
        else:
            nodo.bef = self.rear
            self.rear.next = nodo
            self.rear = nodo
        self.size += 1
        return data

    def imprimirNodo(self, data):
        if isinstance(data, dict):
            return ', '.join(f"{k}: {v}" for k, v in data.items())
        else:
            return str(data)

    def imprimirColaD(self):
        valor = self.front
        resultado = ""
        while valor:
            resultado += self.imprimirNodo(valor.data) + "\n"
            valor = valor.next
        return resultado.strip()

    def imprimirColaIz(self):
        valor = self.rear
        resultado = ""
        while valor:
            resultado += self.imprimirNodo(valor.data) + "\n"
            valor = valor.bef
        return resultado.strip()

# ============================
# Función para cargar inventario.txt
# ============================
def verificar_archivo(archivo='inventario.txt'):
    if not os.path.exists(archivo):
        messagebox.showerror("Error", f"El archivo '{archivo}' no existe.")
        return False
    if os.stat(archivo).st_size == 0:
        messagebox.showerror("Error", f"El archivo '{archivo}' está vacío.")
        return False
    return True

def cargar_inventario(cola, archivo='inventario.txt'):
    if not verificar_archivo(archivo):
        return
    with open(archivo, 'r') as file:
        for line in file:
            data = json.loads(line.strip())
            cola.enColarD(data)

# ============================
# Tkinter: Interfaz gráfica
# ============================
def mostrar_cola(order="FIFO"):
    texto_area.delete(1.0, tk.END)
    if order == "FIFO":
        datos = cola.imprimirColaD()
    else:
        datos = cola.imprimirColaIz()
    texto_area.insert(tk.END, datos if datos else "La cola está vacía.")

def registrar_datos():
    try:
        nuevo_dato = {
            "id": int(entry_id.get()),
            "nombre": entry_nombre.get(),
            "tipo": entry_tipo.get(),
            "categoria": entry_categoria.get(),
            "cantidad": int(entry_cantidad.get()),
            "fech_venc": entry_fecha.get()
        }
        cola.enColarD(nuevo_dato)
        messagebox.showinfo("Registro exitoso", "Producto registrado con éxito.")
        limpiar_campos()
        mostrar_cola()  # Actualizar la lista mostrada
    except ValueError:
        messagebox.showerror("Error de entrada", "Por favor, verifica los datos ingresados.")

def limpiar_campos():
    entry_id.delete(0, tk.END)
    entry_nombre.delete(0, tk.END)
    entry_tipo.delete(0, tk.END)
    entry_categoria.delete(0, tk.END)
    entry_cantidad.delete(0, tk.END)
    entry_fecha.delete(0, tk.END)

# Crear la ventana principal
ventana = tk.Tk()
ventana.title("Gestión de Inventario")
ventana.geometry("600x400")

# Crear la estructura de pestañas
notebook = ttk.Notebook(ventana)
notebook.pack(expand=1, fill="both")

# Pestaña: Visualizar cola
frame_visualizar = ttk.Frame(notebook)
notebook.add(frame_visualizar, text="Visualizar Cola")

texto_area = Text(frame_visualizar, wrap="word", width=70, height=20)
texto_area.pack(pady=10)

boton_fifo = tk.Button(frame_visualizar, text="Mostrar FIFO", command=lambda: mostrar_cola("FIFO"))
boton_fifo.pack(side="left", padx=5, pady=5)

boton_lifo = tk.Button(frame_visualizar, text="Mostrar LIFO", command=lambda: mostrar_cola("LIFO"))
boton_lifo.pack(side="right", padx=5, pady=5)

# Pestaña: Registrar producto
frame_registrar = ttk.Frame(notebook)
notebook.add(frame_registrar, text="Registrar Producto")

tk.Label(frame_registrar, text="ID:").grid(row=0, column=0, padx=5, pady=5)
entry_id = tk.Entry(frame_registrar)
entry_id.grid(row=0, column=1, padx=5, pady=5)

tk.Label(frame_registrar, text="Nombre:").grid(row=1, column=0, padx=5, pady=5)
entry_nombre = tk.Entry(frame_registrar)
entry_nombre.grid(row=1, column=1, padx=5, pady=5)

tk.Label(frame_registrar, text="Tipo:").grid(row=2, column=0, padx=5, pady=5)
entry_tipo = tk.Entry(frame_registrar)
entry_tipo.grid(row=2, column=1, padx=5, pady=5)

tk.Label(frame_registrar, text="Categoría:").grid(row=3, column=0, padx=5, pady=5)
entry_categoria = tk.Entry(frame_registrar)
entry_categoria.grid(row=3, column=1, padx=5, pady=5)

tk.Label(frame_registrar, text="Cantidad:").grid(row=4, column=0, padx=5, pady=5)
entry_cantidad = tk.Entry(frame_registrar)
entry_cantidad.grid(row=4, column=1, padx=5, pady=5)

tk.Label(frame_registrar, text="Fecha de Vencimiento:").grid(row=5, column=0, padx=5, pady=5)
entry_fecha = tk.Entry(frame_registrar)
entry_fecha.grid(row=5, column=1, padx=5, pady=5)

boton_registrar = tk.Button(frame_registrar, text="Registrar", command=registrar_datos)
boton_registrar.grid(row=6, column=0, columnspan=2, pady=10)

# Inicializar cola y cargar inventario
cola = ColaDob()
cargar_inventario(cola)

ventana.mainloop()
```
